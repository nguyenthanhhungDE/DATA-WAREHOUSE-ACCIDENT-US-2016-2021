```python
import pandas as pd
import random
import math
import numpy as np
```

```python
df = pd.read_csv('./Vehicle.csv', delimiter=",",
                 encoding="iso-8859-1", low_memory=False)
```

```python
# Xu ly Age_Band_of_Driver
unique_values = df['Age_Band_of_Driver'].dropna().unique()

# Random các giá trị từ cột
random_values = np.random.choice(
    unique_values, size=df['Age_Band_of_Driver'].isnull().sum())
df['Age_Band_of_Driver'].fillna(value=pd.Series(random_values,
                                                index=df.index[df['Age_Band_of_Driver'].isnull()]), inplace=True)


lst_Age_Band_Of_Driver = list(unique_values)
lst_remove = ['Data missing or out of range',
              '15-Nov', '10-Jun', '0 - 5', 'Not known', 'None']
for value in lst_remove:
    if value in lst_Age_Band_Of_Driver:
        lst_Age_Band_Of_Driver.remove(value)

for i, v in df.iterrows():
    if v['Age_Band_of_Driver'] in lst_remove:
        df.loc[i, 'Age_Band_of_Driver'] = random.choice(lst_Age_Band_Of_Driver)
```

```python
# Xu ly Age_of_Vehicle
lst_Age_of_Vehicle = list(df['Age_of_Vehicle'].dropna().unique())
for i, v in df.iterrows():
    if math.isnan(v['Age_of_Vehicle']):
        df.loc[i, 'Age_of_Vehicle'] = random.choice(lst_Age_of_Vehicle)
```

```python
# Xu ly Driver_Home_Area_Type
unique_values = df['Driver_Home_Area_Type'].dropna().unique()

# Random các giá trị từ cột
random_values = np.random.choice(
    unique_values, size=df['Driver_Home_Area_Type'].isnull().sum())
df['Driver_Home_Area_Type'].fillna(value=pd.Series(random_values,
                                                   index=df.index[df['Driver_Home_Area_Type'].isnull()]), inplace=True)


lst_Driver_Home_Area_Type = list(unique_values)
for value in lst_remove:
    if value in lst_Driver_Home_Area_Type:
        lst_Driver_Home_Area_Type.remove(value)

for i, v in df.iterrows():
    if v['Driver_Home_Area_Type'] in lst_remove:
        df.loc[i, 'Driver_Home_Area_Type'] = random.choice(
            lst_Driver_Home_Area_Type)
```

```python
# Xu ly Sex_of_Driver
unique_values = df['Sex_of_Driver'].dropna().unique()
# Random các giá trị từ cột
random_values = np.random.choice(
    unique_values, size=df['Sex_of_Driver'].isnull().sum())
df['Sex_of_Driver'].fillna(value=pd.Series(random_values,
                                                   index=df.index[df['Sex_of_Driver'].isnull()]), inplace=True)

lst_Sex_of_Driver = list(unique_values)
for value in lst_remove:
    if value in lst_Sex_of_Driver:
        lst_Sex_of_Driver.remove(value)

for i, v in df.iterrows():
    if v['Sex_of_Driver'] in lst_remove:
        df.loc[i, 'Sex_of_Driver'] = random.choice(
            lst_Sex_of_Driver)
```

```python
# Xu ly Junction_Location
unique_values = df['Junction_Location'].dropna().unique()

# Random các giá trị từ cột
random_values = np.random.choice(
    unique_values, size=df['Junction_Location'].isnull().sum())
df['Junction_Location'].fillna(value=pd.Series(random_values,
                                           index=df.index[df['Junction_Location'].isnull()]), inplace=True)

lst_Junction_Location = list(unique_values)
for value in lst_remove:
    if value in lst_Junction_Location:
        lst_Junction_Location.remove(value)

for i, v in df.iterrows():
    if v['Junction_Location'] in lst_remove:
        df.loc[i, 'Junction_Location'] = random.choice(
            lst_Junction_Location)
```

```python
df.to_csv('./Vehicle.csv', index=False)
```
